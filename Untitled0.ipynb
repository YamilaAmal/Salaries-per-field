```python
from scipy.stats import ttest_ind

# Filtrar datos por género
male_salaries = df[df["Gender"] == "Male"]["Starting_Salary"]
female_salaries = df[df["Gender"] == "Female"]["Starting_Salary"]

# Prueba t de Student
t_stat, p_value = ttest_ind(male_salaries, female_salaries, equal_var=False)

# Mostrar resultados
print(f"Estadístico t: {t_stat:.2f}")
print(f"Valor p: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print("✅ Hay una diferencia estadísticamente significativa en los salarios por género.")
else:
    print("❌ No hay evidencia suficiente para decir que los salarios difieren significativamente por género.")
```

```python



# Variable de interés
column = "Starting_Salary"

# Función para calcular outliers por carrera
def detect_outliers_iqr(group):
    Q1 = group[column].quantile(0.25)
    Q3 = group[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group[column] < lower_bound) | (group[column] > upper_bound)]

# Detectar outliers por carrera
outliers_by_career = df.groupby("Field_of_Study", group_keys=False).apply(detect_outliers_iqr)

# Mostrar cuántos outliers hay por carrera
print(outliers_by_career["Field_of_Study"].value_counts())

# Visualizar boxplots por carrera
plt.figure(figsize=(14, 6))
sns.boxplot(x="Field_of_Study", y=column, data=df)
plt.xticks(rotation=45)
plt.title("Distribución de Starting Salary por carrera con outliers")
plt.show()
```

```python


# 📊 MÉTRICAS ESTADÍSTICAS GENERALES
Q1 = df["Starting_Salary"].quantile(0.25)  # Primer cuartil (25%)
Q3 = df["Starting_Salary"].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartílico
lower_bound = Q1 - 1.5 * IQR  # Límite inferior
upper_bound = Q3 + 1.5 * IQR  # Límite superior

# 📏 Outliers detectados por IQR
outliers_iqr = df[(df["Starting_Salary"] < lower_bound) | (df["Starting_Salary"] > upper_bound)]

# 📊 MÉTRICAS BASADAS EN DESVIACIÓN ESTÁNDAR (3 SIGMAS)
mean_salary = df["Starting_Salary"].mean()
std_salary = df["Starting_Salary"].std()
lower_bound_sigma = mean_salary - 3 * std_salary
upper_bound_sigma = mean_salary + 3 * std_salary

# 📏 Outliers detectados por 3 sigmas
outliers_sigma = df[(df["Starting_Salary"] < lower_bound_sigma) | (df["Starting_Salary"] > upper_bound_sigma)]

# 📌 MOSTRAR RESULTADOS
print(f"Salario Mínimo: {df['Starting_Salary'].min():,.2f}")
print(f"Salario Máximo: {df['Starting_Salary'].max():,.2f}")
print(f"Media Salarial: {mean_salary:,.2f}")
print(f"Desviación Estándar: {std_salary:,.2f}")
print(f"Rango Intercuartílico (IQR): {IQR:,.2f}")
print(f"Límite Inferior (IQR): {lower_bound:,.2f}")
print(f"Límite Superior (IQR): {upper_bound:,.2f}")
print(f"Outliers por IQR: {len(outliers_iqr)} registros")
print(f"Outliers por 3 Sigmas: {len(outliers_sigma)} registros")
```

Como hay una gran cantidad de personas que cobra un salario muy bajo, vamos a ver a qué carreras pertenecen

```python
# Definir umbral para salarios bajos ( menores a $30,000)
umbral_bajo = 30000

# Filtrar los datos con salarios por debajo del umbral
salarios_bajos = copydf[df["Starting_Salary"] <= umbral_bajo]

# Contar cuántos registros hay por cada carrera
carreras_salarios_bajos = salarios_bajos["Field_of_Study"].value_counts()

# Mostrar la distribución
print("Distribución de carreras con salarios bajos:")
print(carreras_salarios_bajos)

# Visualizar con un gráfico de barras
plt.figure(figsize=(10,5))
sns.barplot(x=carreras_salarios_bajos.index, y=carreras_salarios_bajos.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.xlabel("Carrera")
plt.ylabel("Cantidad de salarios bajos")
plt.title("Distribución de carreras con salarios bajos (<$30,000)")
plt.show()
```

Podemos sospechar que las carreras mas técnicas necesitan mas experiencia para cobrar mas, en cambio artes no necesitan experiencia. No hay diferencia de salarios según el genero ni hay correlaciones con ninguna de las variables. Vamos a ver qué carreras reciben mas ofertas
