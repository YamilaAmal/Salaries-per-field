```python
from scipy.stats import ttest_ind

# Filtrar datos por gÃ©nero
male_salaries = df[df["Gender"] == "Male"]["Starting_Salary"]
female_salaries = df[df["Gender"] == "Female"]["Starting_Salary"]

# Prueba t de Student
t_stat, p_value = ttest_ind(male_salaries, female_salaries, equal_var=False)

# Mostrar resultados
print(f"EstadÃ­stico t: {t_stat:.2f}")
print(f"Valor p: {p_value:.4f}")

# InterpretaciÃ³n
if p_value < 0.05:
    print("âœ… Hay una diferencia estadÃ­sticamente significativa en los salarios por gÃ©nero.")
else:
    print("âŒ No hay evidencia suficiente para decir que los salarios difieren significativamente por gÃ©nero.")
```

```python



# Variable de interÃ©s
column = "Starting_Salary"

# FunciÃ³n para calcular outliers por carrera
def detect_outliers_iqr(group):
    Q1 = group[column].quantile(0.25)
    Q3 = group[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group[column] < lower_bound) | (group[column] > upper_bound)]

# Detectar outliers por carrera
outliers_by_career = df.groupby("Field_of_Study", group_keys=False).apply(detect_outliers_iqr)

# Mostrar cuÃ¡ntos outliers hay por carrera
print(outliers_by_career["Field_of_Study"].value_counts())

# Visualizar boxplots por carrera
plt.figure(figsize=(14, 6))
sns.boxplot(x="Field_of_Study", y=column, data=df)
plt.xticks(rotation=45)
plt.title("DistribuciÃ³n de Starting Salary por carrera con outliers")
plt.show()
```

```python


# ðŸ“Š MÃ‰TRICAS ESTADÃSTICAS GENERALES
Q1 = df["Starting_Salary"].quantile(0.25)  # Primer cuartil (25%)
Q3 = df["Starting_Salary"].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartÃ­lico
lower_bound = Q1 - 1.5 * IQR  # LÃ­mite inferior
upper_bound = Q3 + 1.5 * IQR  # LÃ­mite superior

# ðŸ“ Outliers detectados por IQR
outliers_iqr = df[(df["Starting_Salary"] < lower_bound) | (df["Starting_Salary"] > upper_bound)]

# ðŸ“Š MÃ‰TRICAS BASADAS EN DESVIACIÃ“N ESTÃNDAR (3 SIGMAS)
mean_salary = df["Starting_Salary"].mean()
std_salary = df["Starting_Salary"].std()
lower_bound_sigma = mean_salary - 3 * std_salary
upper_bound_sigma = mean_salary + 3 * std_salary

# ðŸ“ Outliers detectados por 3 sigmas
outliers_sigma = df[(df["Starting_Salary"] < lower_bound_sigma) | (df["Starting_Salary"] > upper_bound_sigma)]

# ðŸ“Œ MOSTRAR RESULTADOS
print(f"Salario MÃ­nimo: {df['Starting_Salary'].min():,.2f}")
print(f"Salario MÃ¡ximo: {df['Starting_Salary'].max():,.2f}")
print(f"Media Salarial: {mean_salary:,.2f}")
print(f"DesviaciÃ³n EstÃ¡ndar: {std_salary:,.2f}")
print(f"Rango IntercuartÃ­lico (IQR): {IQR:,.2f}")
print(f"LÃ­mite Inferior (IQR): {lower_bound:,.2f}")
print(f"LÃ­mite Superior (IQR): {upper_bound:,.2f}")
print(f"Outliers por IQR: {len(outliers_iqr)} registros")
print(f"Outliers por 3 Sigmas: {len(outliers_sigma)} registros")
```
